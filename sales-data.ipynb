```python
# 3. How is the data distributed across different invoice dates? Are there any seasonal trends or patterns? (Use a graph for this.)'''
```

```python
import pandas as pd
df5 = pd.read_csv('fct_invoice.csv')

df5
```

```python
df5['invoice_date'] = pd.to_datetime(df5['invoice_date'])

df5['invoice_month'] = df5['invoice_date'].dt.month
df5['invoice_year'] = df5['invoice_date'].dt.year

df5
```

```python
month = df5['invoice_month'].unique()
month.sort()
months = []
for i in range(len(month)):
    months.append(month[i])
    
months
```

```python
year = df5['invoice_year'].unique()
year.sort()

years = []
for i in range(len(year)):
     years.append(year[i])

years
```

```python
df6 = df5.groupby(["invoice_year","invoice_month"]).count()
df6
```

```python
count = df6['invoice_no'].values.tolist()
print(count)
```

```python
chunk_size = 12

split_lists = [count[i:i+chunk_size] for i in range(0, len(count), chunk_size)]
# used chatgpt to learn how to split the big list into 3 lists of 12 to represent the months in the 3 years (2021,2022,2023)

print(split_lists)
```

```python
year2021 = split_lists[0]
year2022 = split_lists[1]
year2023 = split_lists[2]
```

```python
import matplotlib.pyplot as plt

plt.plot(months,year2021)

plt.plot(months,year2021, label = "2021")
plt.plot(months,year2022, label = "2022")
plt.plot(months,year2023, label = "2023")

plt.title("Seasonality Trends (Purchases per Month)")
plt.xlabel("Months")
plt.ylabel("Number of Purchases")

plt.xticks(range(1,len(months)+1))

plt.grid(True)
plt.legend()
plt.show()
```

Medium #3: Seasonality Trends

Descriptions
- The month that records the highest number of purchases is July (2021).
- From January, a peak along with December, the purchases started to dip in February as the holiday season ended.
- After February, the purchases began to peak in March. 
- Purchases were did not see a big change within the months April to June, until the number starts to increase during the summer time (July and August).
- Prior to the holidays, purchases also seem to increase on October.

Analysis
- The months that record high numbers of purchases are the following: December and January, March, July and August, October. These seem to be months wherein students have no school due to breaks.
    - December and January: holiday season
    - July and August: summer season
    - March and October: possibly semestral breaks
- The seasonality trends for 2021 and 2022 followed a similar pattern. Whereas the number of purchases in 2023 saw a substantial drop, and the year follows a completely different trend than the previous two years, possibly making it an outlier.

```python
# creating age-classes
# the code was from chatgpt
bins = range(0,100,10)
# sets range of values from 0-100, third value means intervals of 10

labels = [f'{start}-{end-1}' for start, end in zip(bins, bins[1:])]


df3['age_class'] = pd.cut(df3['age'], bins=bins, labels=labels, right=False)
# pd.cut() gets the values from the panda and checks which class it belongs to

df3
```

```python
pivot_table = df3.pivot_table(index='age_class', columns='category', values='total_sales', aggfunc='sum')
pivot_table
```
