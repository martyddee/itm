```python
# converting data type from str to float
latitudes_list = [float(element) for element in latitudes]
longitudes_list = [float(element) for element in longitudes]
```

```python
plt.scatter(latitudes_list,longitudes_list)
plt.title("Location of the International Space Station (ISS)")
plt.xlabel("latitude")
plt.ylabel("longitude")
```

Insights
- In the path of the ISS, it seemed to travel in a logarithmic line.
- At around (18, -112), the ISS starts to travel in an exponential line.

```python
plt.plot(timestamps,latitudes_list)
plt.title("this is a graph")
plt.xlabel("timestamp")
plt.ylabel("latitude/longitude")

plt.plot(timestamps,longitudes_list)

plt.plot(timestamps,latitudes_list, label = "change in latitudes")
plt.plot(timestamps,longitudes_list, label = "change in longitudes")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

fig, ax = plt.subplots(1,2, figsize = [11,5])

ax[0].scatter(timestamps,latitudes_list, color = "green")
ax[0].set_title("Change in the ISS' Latitudes")

ax[1].scatter(timestamps,longitudes_list, color = "red")
ax[1].set_title("Change in the ISS' Longitudes")

plt.show()
```

Insights
- the change of latitudes seems to a follow a logarithmic trend
- the change of longitudes seems to a follow a exponential trend

```python
import requests
import json
import time

page = requests.get("http://api.open-notify.org/iss-now.json")
time.sleep(10)
# added time.sleep(10) to implement the 10-second intervals
info = page.json()

information.append(info)

print(information)
print(len(information))
```

```python
latitudes = []
for i in range(len(information)):
    latitudes.append(information[i]['iss_position']['latitude'])
    
latitudes
```

```python
latitudes_list = [float(element) for element in latitudes]

latitudes_list
```

```python
longitudes = []
for i in range(len(information)):
    longitudes.append(information[i]['iss_position']['longitude'])
    
longitudes
```

```python
longitudes_list = [float(element) for element in longitudes]

longitudes_list
```

```python
timestamps = []
for i in range(len(information)):
    timestamps.append(information[i]['timestamp'])
    
timestamps
```

```python
import matplotlib.pyplot as plt

plt.scatter(latitudes_list,longitudes_list)
plt.title("Location of the International Space Station (ISS)")
plt.xlabel("latitude")
plt.ylabel("longitude")
```

Insights
- In the path of the ISS, it seemed to travel in a logarithmic line from the first point (-50,-40) to the coordinate (0,35).
- From (0,35) to the last point (28,57), the ISS starts to travel in an exponential line.

```python
plt.plot(timestamps,latitudes_list)
plt.title("this is a graph")
plt.xlabel("timestamp")
plt.ylabel("latitude/longitude")

plt.plot(timestamps,longitudes_list)

plt.plot(timestamps,latitudes_list, label = "change in latitudes")
plt.plot(timestamps,longitudes_list, label = "change in longitudes")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

fig, ax = plt.subplots(1,2, figsize = [11,5])

ax[0].scatter(timestamps,latitudes_list, color = "green")
ax[0].set_title("Change in the ISS' Latitudes")

ax[1].scatter(timestamps,longitudes_list, color = "red")
ax[1].set_title("Change in the ISS' Longitudes")

plt.show()
```

Insights
- the change of latitudes seems to a follow a exponential trend throughout all the data points gathered
- the change of longitudes seems to a follow a logarithmic trend until the point (2000,32) where it start to follow a exponential trend
